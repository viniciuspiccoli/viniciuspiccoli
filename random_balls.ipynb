```python

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter

num_spheres = 50
size = 100

positions = np.random.rand(num_spheres, 2) * size
velocities = (np.random.rand(num_spheres, 2) - 0.5) * 2

fig, ax = plt.subplots()
spheres = []

for _ in range(num_spheres):
    sphere = plt.Circle((0, 0), 2, fc='blue')
    ax.add_patch(sphere)
    spheres.append(sphere)

ax.set_xlim(0, size)
ax.set_ylim(0, size)

def update(frame):
    for i, sphere in enumerate(spheres):
        positions[i] += velocities[i]
        if positions[i][0] > size or positions[i][0] < 0:
            velocities[i][0] = -velocities[i][0]
        if positions[i][1] > size or positions[i][1] < 0:
            velocities[i][1] = -velocities[i][1]
        sphere.center = positions[i]
    return spheres

ani = animation.FuncAnimation(fig, update, frames=1000, interval=1, blit=True)

# Save the animation
ani.save('md_simulation.gif', writer=PillowWriter(fps=60))

plt.axis('off')
plt.show()
```
